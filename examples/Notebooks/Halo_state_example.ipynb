```python
import cse
import matplotlib.pyplot as plt
import numpy as np
sr2_halo = cse.Cse('86Sr', VT=['../potentials/srGS_largePosScat.dat'])
```

```python
sr2_halo.solve(sr2_halo.VT[0,0,-1] - 150*4.13567*10**-9)
sr2_halo
```

```python
plt.plot(sr2_halo.R, sr2_halo.VT[0, 0], label='dia. E')
#plt.plot(sr2_halo.R, 0.01*sr2_halo.wavefunction[:, 0] + sr2_halo.energy)
plt.axis(xmin=0, xmax=200, ymin=.13, ymax=.14)
plt.show()
```

```python
(sr2_halo.VT[0,0,-1] - sr2_halo.energy)*2.41799*10**8
```

```python
wfC = 4*np.pi
wfPhi = np.power(np.absolute(sr2_halo.wavefunction[:, 0]), 2)
wf = np.multiply(wfC, np.transpose(wfPhi))
                 
plt.plot(sr2_halo.R, sr2_halo.VT[0, 0], label='dia. E')
#plt.plot(sr2_halo.R, np.power(np.absolute(sr2_halo.wavefunction[:, 0]),2) + sr2_halo.energy)
plt.plot(sr2_halo.R, 0.1*wf[0] + sr2_halo.energy)
plt.axis(xmin=0, xmax=200, ymin=0.134, ymax=0.18)
plt.show()
```
