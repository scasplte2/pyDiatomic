```python
import cse
import matplotlib.pyplot as plt
import numpy as np

ev2Ha = 3.67493e-2
cm2Ha = 4.55634e-6
cm2MHz = 2.99792e4
ev2cm = 8065.541
ev2MHz = 2.41799e8
aBohr = 0.52917721067

sr0u = cse.Cse('86Sr', VT=['../potentials/srVR.dat'])
sr1u = cse.Cse('86Sr', VT=['../potentials/srVF.dat'])

sr0u.VT[0,0] = (sr0u.VT[0,0] - sr1u.VT[0,0].min())/ev2cm # align their asymptotes
sr1u.VT[0,0] = (sr1u.VT[0,0] - sr1u.VT[0,0].min())/ev2cm
```

```python
sr1u.solve(sr1u.VT[0,0,-1] - 2828/ev2MHz)
wf0 = sr1u.wavefunction[:, 0]
en0 = sr1u.energy
(en0 - sr1u.VT[0,0,-1])*ev2MHz
```

```python
sr0u.solve(sr0u.VT[0,0,-1] - 10/ev2MHz)
wf1 = sr0u.wavefunction[:, 0]
en1 = sr0u.energy
(en1 - sr0u.VT[0,0,-1])*ev2MHz
```

```python
sr0u.solve(sr0u.VT[0,0,-1] - 44.233*1/ev2MHz)
wf2 = sr0u.wavefunction[:, 0]
en2 = sr0u.energy
(en2 - sr0u.VT[0,0,-1])*ev2MHz
```

```python
sr0u.solve(sr0u.VT[0,0,-1] - 348.73*1/ev2MHz)
wf3 = sr0u.wavefunction[:, 0]
en3 = sr0u.energy
(en3 - sr0u.VT[0,0,-1])*ev2MHz
```

```python
plt.figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')

# Potential to plots
V0u = (sr0u.VT[0, 0] - sr0u.VT[0, 0, -1])*ev2cm
V1u = (sr1u.VT[0,0] - sr1u.VT[0, 0, -1])*ev2cm

# length
R = sr0u.R/aBohr;

# Wavefunction to plot
A = 400 # arbitrary amplitude factor for visualization
#wf1u_1 = A*wf0 + (en0 - sr1u.VT[0, 0, -1])*ev2MHz
#wf0u_1 = A*wf1 + (en1 - sr0u.VT[0, 0, -1])*ev2MHz
#wf0u_1 = A*wf2 + (en2 - sr0u.VT[0, 0, -1])*ev2MHz
#wf0u_1 = A*wf3 + (en3 - sr0u.VT[0, 0, -1])*ev2MHz

plt.plot(R, V0u, label='$0_u^+$')
plt.plot(R, V1u, label='$1_u$')
#plt.plot(R, wf1u_1)
#plt.plot(R, wf0u_1)

#plt.axis(xmin=5, xmax=250/aBohr, ymin=-500, ymax=50)
plt.axis(xmin=5, xmax=25, ymin=-4000, ymax=2000)
#plt.xscale('log')

plt.xlabel('Interparticle distance, r [a0]')
plt.ylabel('Potential Energy, [MHz]')

plt.rcParams.update({'font.size': 20})
plt.legend(); plt.grid(1); plt.show()
```
